```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train, y_train), (X_test , y_test)=keras.datasets.mnist.load_data()
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train
```

```python
X_train[0]
```

```python
model=Sequential()
#this will convert 2D array int 1D array
model.add(Flatten(input_shape=(28,28)))
#now adding layers in our model
model.add(Dense(128, activation='relu'))        #first layer
model.add(Dense(32, activation='relu'))         #second layer
model.add(Dense(10, activation='softmax'))        #output layer since it is a 10 classification problem
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history=model.fit(X_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_prob=model.predict(X_test)
```

```python
y_prob
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
